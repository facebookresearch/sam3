```python
# Step 1: First authenticate with YouTube using the new headless approach
# Run this to get authenticated cookies

# Option 1: Run authentication with GUI (easier)
import subprocess
import sys

print("Starting YouTube authentication...")
print("This will open a GUI browser window for you to login to YouTube")
print("The browser will be Firefox managed by Playwright")

try:
    result = subprocess.run([
        sys.executable, "youtube_auth.py", 
        "--gui",  # Show browser window
        "--interactive"  # Wait for manual login
    ], capture_output=True, text=True, timeout=600)  # 10 minute timeout
    
    if result.returncode == 0:
        print("✓ Authentication successful!")
        print("Cookies saved to youtube_cookies.txt")
    else:
        print("✗ Authentication failed:")
        print("STDOUT:", result.stdout)
        print("STDERR:", result.stderr)
        
except subprocess.TimeoutExpired:
    print("Authentication timed out. Please try again.")
except Exception as e:
    print(f"Error during authentication: {e}")
```
