```python
with open("/fsx-onevision/tym/sam3_and_data/data/pred/haithamkhedr_sam3_dense_sam3_v2_sam3_v2_rc1_eval_saco_veval_hot_recon16_MF_yaml_0_preds/saco_veval_sav_val_preds.json.sam3_eval_metrics", "r") as f:
    curr_report = json.load(f)
    curr_report = curr_report["dataset_level_metrics"]

def update_k(k, dataset_name):
    if k.startswith(dataset_name):
        k = k.removeprefix(dataset_name)
        k = "video" + k
    elif k.startswith("ytvis"):
        k = k.replace("ytvis", "video")
    else:
        pass

    return k

dataset_name = "saco_veval_sav_val"
curr_report = {
    update_k(k, dataset_name): v for k, v in curr_report.items()
}
```

```python
match_res = []
for curr_k, curr_v in curr_report.items():
    curr_v = float(curr_v)
    val_match_k = []
    val_match_v = []
    match_reason = []
    for new_k, new_v in new_report.items():
        match_reason_per_v = []
        new_v = float(new_v)

        if (curr_k == new_k):
            val_match_k.append(new_k)
            val_match_v.append(new_v)
            
            if curr_v == new_v:
                match_reason_per_v.append("value_match")
            if curr_k == new_k:
                match_reason_per_v.append("metric_name_match")
            match_reason.append(match_reason_per_v)

    match_res_per_match = {
        "curr_k": curr_k,
        "curr_v": curr_v,
        "new_val_match_k": val_match_k,
        "new_val_match_v": val_match_v,
        "match_reason": match_reason,
    }
    match_res.append(match_res_per_match)

match_res_df = pd.DataFrame(match_res)
match_res_df
```
